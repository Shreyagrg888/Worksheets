```python
#Task 3 outlier detection

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Calculation of IQR for score and GDP
Q1_score = dataset['score'].quantile(0.25)
Q3_score = dataset['score'].quantile(0.75)
IQR_score = Q3_score - Q1_score

Q1_gdp = dataset['Log GDP per capita'].quantile(0.25)
Q3_gdp = dataset['Log GDP per capita'].quantile(0.75)
IQR_gdp = Q3_gdp - Q1_gdp

# Lower bound and upper bound using 1.5 * IQR rule
lower_bound_score = Q1_score - 1.5 * IQR_score
upper_bound_score = Q3_score + 1.5 * IQR_score

lower_bound_gdp = Q1_gdp - 1.5 * IQR_gdp
upper_bound_gdp = Q3_gdp + 1.5 * IQR_gdp

# 2. Identifying outliers
outlier_score = dataset[(dataset['score'] < lower_bound_score) | (dataset['score'] > upper_bound_score)]
outlier_gdp = dataset[(dataset['Log GDP per capita'] < lower_bound_gdp) | (dataset['Log GDP per capita'] > upper_bound_gdp)]

# Combining the outliers from both score and GDP
outliers = pd.concat([outlier_score, outlier_gdp]).drop_duplicates()

# 3. Making the scatter plot
plt.figure(figsize=(10, 5))

# Plot the entire dataset first (in blue)
sns.scatterplot(x='Log GDP per capita', y='score', data=dataset, color='blue', label='Normal')

# Overlay the outliers on top (in red), without overriding the normal data
sns.scatterplot(x='Log GDP per capita', y='score', data=outliers, color='red', label='Outliers')

# Adding the title and labels
plt.title('Scatter plot of GDP per Capita vs Score with Outliers Highlighted')
plt.xlabel('Log GDP per Capita')
plt.ylabel('Score')

# Show the legend
plt.legend()

# Display the plot
plt.show()

```
